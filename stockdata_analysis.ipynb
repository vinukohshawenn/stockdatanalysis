```python
import yfinance as yf
import pandas as pd

tesla = yf.Ticker("TSLA")
tesla_financials = tesla.financials

tesla_revenue = tesla_financials.transpose()
tesla_revenue.reset_index(inplace=True)

tesla_revenue = tesla_revenue.rename(columns = {'index':'Date'})
print(tesla_revenue[['Date', 'Total Revenue']].tail())
```

Question 05:

visualizing the given Tesla's stock data.

```python
import yfinance as yf
import matplotlib.pyplot as plt

tesla_data = yf.Ticker("TSLA")
tesla_df = tesla_data.history(period="max")

def make_graph(data, title):
    plt.figure(figsize=(12,6))
    plt.plot(data.index, data['Close'], label='Closing Price')
    plt.title(title)
    plt.xlabel("Date")
    plt.ylabel("Closing Price (USD)")
    plt.grid(True)
    plt.legend()
    plt.show()
make_graph(tesla_df, "Tesla Stock Price Over Time")
```
